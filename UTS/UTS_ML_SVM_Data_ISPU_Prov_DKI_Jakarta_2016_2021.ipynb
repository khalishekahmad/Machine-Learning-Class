# **Konsep-konsep Machine Learning**

Internship-Robby-Cabang-AI.jpg



Machine learning (ML) adalah cabang dari kecerdasan buatan yang fokus pada pengembangan sistem yang dapat belajar dari data. Tujuan utama dari machine learning adalah memungkinkan komputer untuk belajar dan membuat keputusan atau tugas tanpa diprogram secara eksplisit. Ada beberapa konsep dasar yang perlu dipahami dalam machine learning:

1. **Data:**
   - **Training Data:** Merupakan data yang digunakan untuk melatih model. Data ini harus mencakup contoh-contoh yang representatif dari situasi atau masalah yang ingin dipecahkan.
   - **Testing Data:** Data yang digunakan untuk menguji kinerja model setelah dilatih. Data ini seharusnya tidak dikenal oleh model selama proses pelatihan.

2. **Model:**
   - **Algoritma:** Sebuah set prosedur matematis yang digunakan oleh model untuk membuat prediksi atau keputusan berdasarkan data input.
   - **Parameter:** Nilai-nilai yang disesuaikan oleh model selama proses pelatihan untuk meningkatkan kemampuannya membuat prediksi yang akurat.

3. **Feature (Fitur):**
   - **Variable Input:** Data yang digunakan sebagai input untuk model. Fitur-fitur ini dapat menjadi atribut atau karakteristik dari data yang diukur atau diamati.

4. **Label atau Target:**
   - **Variable Output:** Hasil yang diharapkan dari model. Dalam pembelajaran terawasi, ini sering disebut sebagai label atau target.

5. **Pelatihan (Training):**
   - **Proses Pengoptimalan:** Model diperbarui atau disesuaikan menggunakan algoritma pelatihan dengan menggunakan data pelatihan. Tujuan utama adalah mengurangi kesalahan model terhadap data pelatihan dan meningkatkan kinerjanya terhadap data yang tidak dikenal.

6. **Validasi dan Pengujian:**
   - **Validasi:** Penggunaan subset data yang tidak digunakan selama pelatihan untuk mengevaluasi kinerja model dan mencegah overfitting.
   - **Pengujian (Testing):** Penggunaan data yang sepenuhnya tidak dikenal oleh model untuk mengukur sejauh mana model dapat membuat prediksi yang akurat.

7. **Supervised, Unsupervised, dan Reinforcement Learning:**
   - **Supervised Learning:** Model belajar dari pasangan input-output yang sudah diberi label.
   - **Unsupervised Learning:** Model belajar dari data yang tidak diberi label dan mencoba menemukan pola atau struktur di dalamnya.
   - **Reinforcement Learning:** Model belajar dari interaksi dengan lingkungannya, menerima umpan balik berdasarkan tindakan yang diambil.

8. **Overfitting dan Underfitting:**
   - **Overfitting:** Model terlalu rumit dan "memorizing" data pelatihan sehingga tidak dapat umum pada data baru.
   - **Underfitting:** Model terlalu sederhana sehingga gagal menangkap pola dalam data pelatihan dan tidak dapat melakukan prediksi yang baik pada data baru.

Konsep-konsep ini mencakup dasar-dasar machine learning, dan ada berbagai algoritma dan teknik yang dapat diterapkan berdasarkan jenis tugas dan jenis data yang digunakan. Machine learning telah menjadi alat yang sangat penting dalam berbagai aplikasi, seperti pengenalan wajah, penerjemahan bahasa, analisis data besar, dan banyak lagi.

Konsep dasar cara kerja Machine Learning dapat diuraikan sebagai berikut:

### 1. **Data Gathering:**
   - Machine Learning memerlukan data sebagai bahan bakar untuk pembelajaran. Data ini dapat berupa berbagai jenis informasi tergantung pada jenis masalah yang ingin dipecahkan.

### 2. **Data Preprocessing:**
   - Data yang dikumpulkan seringkali memerlukan pembersihan dan penyiapan sebelum dapat digunakan untuk pelatihan model. Proses ini melibatkan pemrosesan data untuk menghilangkan kecacatan, mengisi nilai yang hilang, atau mengonversi data ke dalam format yang dapat digunakan.

### 3. **Feature Extraction:**
   - Fitur-fitur atau atribut dari data yang relevan untuk masalah harus diekstraksi. Ini melibatkan pemilihan informasi yang paling penting untuk melatih model.

### 4. **Model Selection:**
   - Pemilihan model atau algoritma ML yang sesuai berdasarkan jenis masalah yang dihadapi. Ini bisa mencakup model-model seperti regresi, pohon keputusan, atau jaringan saraf tiruan.

### 5. **Training the Model:**
   - Model ML dilatih dengan memberikan data latihan yang memungkinkan algoritma untuk menyesuaikan parameter internalnya. Selama pelatihan, model mencoba untuk memahami pola dan hubungan dalam data.

### 6. **Evaluation:**
   - Setelah pelatihan, model dievaluasi menggunakan data yang tidak digunakan sebelumnya (data uji) untuk mengukur kinerjanya. Metrik evaluasi dapat berbeda-beda tergantung pada jenis masalah (misalnya, akurasi untuk klasifikasi, MSE untuk regresi).

### 7. **Fine-Tuning:**
   - Berdasarkan hasil evaluasi, model mungkin perlu disesuaikan (fine-tuning) untuk meningkatkan kinerjanya. Ini melibatkan penyesuaian parameter model atau mengubah konfigurasi algoritma.

### 8. **Prediction/Inference:**
   - Setelah model dianggap memadai, ia dapat digunakan untuk membuat prediksi atau inferensi pada data baru yang tidak pernah dilihat sebelumnya.

### 9. **Deployment:**
   - Model yang dilatih dan dievaluasi dapat diimplementasikan ke dalam lingkungan produksi, di mana ia dapat digunakan untuk memproses data secara real-time atau batch.

### 10. **Monitoring dan Maintenance:**
    - Model perlu dipantau secara berkala untuk memastikan bahwa kinerjanya tetap optimal. Jika ada perubahan dalam data atau kebutuhan bisnis, model mungkin perlu diupdate atau disesuaikan.

Dalam keseluruhan, konsep utama di balik Machine Learning adalah kemampuan model untuk belajar dari data, memahami pola, dan membuat prediksi atau keputusan tanpa instruksi eksplisit. Proses ini menciptakan suatu siklus iteratif di mana model terus-menerus diperbarui dan disempurnakan seiring berjalannya waktu.

# **EDA**

**EDA (Exploratory Data Analysis)** dalam machine learning merujuk pada proses pemeriksaan dan pemahaman awal terhadap dataset yang akan digunakan. Tujuan dari EDA adalah untuk mendapatkan wawasan yang mendalam tentang struktur dan karakteristik data sebelum menerapkan model machine learning. Berikut adalah beberapa poin kunci tentang EDA:

1. **Pemahaman Data:**
   - EDA membantu untuk memahami struktur dataset, termasuk jumlah dan tipe variabel, serta nilai-nilai yang ada dalam setiap variabel.

2. **Pendeteksian Anomali dan Outlier:**
   - EDA membantu dalam mendeteksi anomali atau outlier yang mungkin memengaruhi hasil model. Pemahaman ini penting untuk membersihkan data sebelum melatih model.

3. **Distribusi Variabel:**
   - EDA memungkinkan pengamatan terhadap distribusi variabel, membantu pengguna untuk memilih transformasi atau normalisasi yang tepat untuk data.

4. **Korelasi antar Variabel:**
   - Analisis korelasi membantu mengidentifikasi hubungan antara variabel, yang dapat memberikan wawasan tentang fitur yang mungkin memiliki dampak besar pada prediksi.

5. **Visualisasi Data:**
   - EDA melibatkan penggunaan visualisasi data seperti histogram, box plot, scatter plot, dan lainnya untuk mengidentifikasi pola, tren, atau kelainan dalam data.

6. **Pemilihan Fitur:**
   - EDA membantu dalam pemilihan fitur dengan memahami hubungan antar fitur dan dampaknya terhadap target. Fitur yang kurang relevan atau saling terkait dapat diidentifikasi.

7. **Persiapan Data:**
   - Melalui EDA, mungkin diperlukan langkah-langkah persiapan data seperti penanganan missing values, normalisasi, atau transformasi variabel.

EDA memberikan landasan yang kuat sebelum membangun model machine learning, membantu menghindari keputusan yang salah atau pengaturan model yang tidak tepat. Dengan memahami karakteristik data, praktisi machine learning dapat membuat keputusan yang lebih cerdas dan mendapatkan hasil yang lebih baik dari model yang dikembangkan.

Apa itu korelasi?

Korelasi adalah ukuran statistik yang menunjukkan sejauh mana dua variabel bergerak bersama-sama. Dengan kata lain, korelasi mengukur sejauh mana perubahan dalam satu variabel berkaitan dengan perubahan dalam variabel lainnya.

Ada tiga jenis korelasi utama:

1. **Korelasi Positif**: Jika satu variabel meningkat, variabel lainnya juga meningkat. Misalnya, tinggi dan berat badan biasanya memiliki korelasi positif; orang yang lebih tinggi cenderung memiliki berat badan yang lebih besar.

2. **Korelasi Negatif**: Jika satu variabel meningkat, variabel lainnya menurun. Misalnya, jumlah jam belajar dan jumlah kesalahan dalam tes biasanya memiliki korelasi negatif; semakin banyak seseorang belajar, semakin sedikit kesalahan yang mereka buat.

3. **Tidak Ada Korelasi**: Tidak ada hubungan yang jelas antara perubahan dalam satu variabel dan perubahan dalam variabel lainnya.

Nilai korelasi biasanya berkisar antara -1 dan 1. Nilai -1 menunjukkan korelasi negatif sempurna, nilai 0 menunjukkan tidak ada korelasi, dan nilai 1 menunjukkan korelasi positif sempurna. Korelasi adalah alat yang sangat berguna dalam statistik dan analisis data.

Bagaimana cara membaca matriks korelasi?

Matriks korelasi adalah tabel yang menunjukkan koefisien korelasi antara setiap pasangan variabel dalam dataset. Nilai dalam matriks berkisar antara -1 hingga 1, dengan -1 menunjukkan korelasi negatif, 0 menunjukkan tidak ada korelasi, dan 1 menunjukkan korelasi positif.

Berikut adalah cara membaca matriks korelasi:

1. **Identifikasi Variabel**: Variabel biasanya ditampilkan di sumbu x dan y dari matriks. Setiap sel dalam matriks mewakili hubungan antara dua variabel.

2. **Baca Nilai Korelasi**: Nilai dalam setiap sel mewakili koefisien korelasi antara dua variabel. Nilai ini berkisar antara -1 dan 1.
   - Nilai **1** berarti ada korelasi positif sempurna antara dua variabel (ketika satu variabel meningkat, variabel lainnya juga meningkat).
   - Nilai **-1** berarti ada korelasi negatif sempurna (ketika satu variabel meningkat, variabel lainnya menurun).
   - Nilai **0** berarti tidak ada korelasi antara variabel.

3. **Perhatikan Warna**: Dalam heatmap, korelasi positif biasanya ditandai dengan warna hangat seperti merah, dan korelasi negatif ditandai dengan warna dingin seperti biru. Warna yang lebih gelap atau lebih terang menunjukkan kekuatan korelasi.

4. **Diagonal Utama**: Diagonal utama dari matriks (dari kiri atas ke kanan bawah) biasanya memiliki nilai 1, karena ini adalah korelasi variabel dengan dirinya sendiri.

Dengan memahami matriks korelasi, kita dapat mengidentifikasi hubungan antara variabel dalam dataset, yang dapat membantu dalam analisis data dan pemilihan fitur.

Dengan kata lain, data visualization mengubah kumpulan data menjadi hal lebih sederhana untuk ditampilkan. Dengan menggunakan elemen visual tersebut, pembaca akan lebih mudah memahami tren, outliers, dan pola dalam suatu dataÂ¹. Dalam bisnis, data visualization memungkinkan para pembuat keputusan untuk melihat analitik yang disajikan secara visual, sehingga mereka dapat memahami konsep yang sulit atau mengidentifikasi pola baru.

Sedangkan **Outlier** atau pencilan adalah data yang nilainya terlalu jauh dengan data lainnyaÂ². Keberadaan outlier akan membuat analisis terhadap data tersebut menjadi bias atau dengan kata lain tidak mencerminkan fenomena yang sebenarnyaÂ². Outlier ini bisa juga dikenal dengan anomali ini adalah sebuah data atau observasi yang menyimpang secara ekstrim dari rata-rata sekumpulan data yang ada. Penyimpangan ini bisa berupa angkanya terlalu tinggi atau terlalu rendah.

Dalam konteks visualisasi data, outlier ini dapat terlihat sebagai titik data atau elemen visual yang jauh dari kumpulan data lainnya. Dengan memvisualisasikan data, kita dapat lebih mudah mengidentifikasi dan memahami outlier ini. Selain itu, visualisasi data juga membantu kita untuk memahami pola, tren, hubungan, dan outlier yang tersembunyi di dalam data yang besar dan kompleks.



Outlier dalam konteks data visualization mengacu pada nilai yang secara signifikan berbeda dari sebagian besar data. Outlier bisa menjadi titik data yang jauh dari sebagian besar observasi atau dapat mewakili anomali atau kesalahan dalam pengukuran atau pengumpulan data.

Konsep data visualization outlier mencakup cara-cara untuk mengidentifikasi, menyoroti, dan memahami outlier dalam data menggunakan visualisasi grafis. Beberapa teknik umum yang digunakan untuk menangani outlier dalam visualisasi data melibatkan penggunaan grafik seperti:

1. **Box Plot (Diagram Kotak):** Box plot dapat memberikan gambaran visual tentang sebaran data dan dengan mudah menyoroti nilai-nilai yang berada jauh dari kuartil.

2. **Scatter Plot (Diagram Sebar):** Scatter plot adalah grafik yang menunjukkan hubungan antara dua variabel. Outlier dapat ditemukan sebagai titik-titik yang jauh dari pola umum pada scatter plot.

3. **Histogram:** Histogram dapat membantu menggambarkan distribusi data, dan outlier dapat terlihat sebagai puncak atau ekor yang mencolok di luar distribusi utama.

4. **Heatmaps:** Heatmap dapat digunakan untuk memvisualisasikan data dalam bentuk matriks warna. Outlier dapat ditemukan sebagai nilai yang menonjol dengan warna yang berbeda.

5. **Violin Plot:** Violin plot merupakan gabungan antara box plot dan kernel density plot. Ini membantu dalam menunjukkan sebaran data dan juga memperlihatkan kepadatan data pada berbagai nilai.

6. **Lorenz Curve:** Ini digunakan terutama dalam ekonomi untuk menunjukkan distribusi kumulatif dari sebuah variabel. Outlier dapat terlihat sebagai ketidaksesuaian besar dari kurva Lorenz yang diharapkan.

Penting untuk dicatat bahwa tidak semua nilai yang dianggap sebagai outlier harus dihapus. Outlier dapat memberikan informasi berharga tentang data, seperti tren yang tidak terduga atau kesalahan pengukuran. Penting untuk memahami konteks data dan memberikan penilaian yang tepat terhadap nilai-nilai outlier tersebut.

# Data Transformation

Data transformation adalah proses mengubah data dari satu bentuk atau format ke bentuk atau format yang lain. Tujuan utama dari data transformation adalah membuat data lebih sesuai untuk analisis atau pengolahan selanjutnya. Proses ini dapat melibatkan beberapa tindakan, seperti penggabungan data, pemfilteran, pemformatan ulang, penghitungan agregat, dan lain sebagainya.

Berikut adalah beberapa konsep utama dalam data transformation:

1. **Pembersihan Data (Data Cleaning):** Proses ini melibatkan identifikasi dan penanganan nilai yang hilang, duplikat, atau tidak valid dalam dataset. Pembersihan data penting untuk memastikan bahwa hasil analisis atau pemodelan didasarkan pada data yang akurat dan dapat diandalkan.

2. **Pemfilteran (Filtering):** Pemfilteran melibatkan pemilihan subset dari data berdasarkan kriteria tertentu. Ini dapat melibatkan penghapusan baris atau kolom tertentu yang tidak diperlukan untuk analisis tertentu.

3. **Penggabungan Data (Data Merging):** Kadang-kadang, data yang relevan tersebar di beberapa sumber atau tabel. Penggabungan data adalah proses menggabungkan data dari beberapa sumber berbeda berdasarkan kunci atau kolom tertentu.

4. **Pemformatan Ulang (Reshaping):** Pemformatan ulang data melibatkan perubahan struktur data. Contohnya, mengubah data dari format panjang menjadi format lebar atau sebaliknya.

5. **Penghitungan Agregat (Aggregation):** Agregasi melibatkan penghitungan statistik ringkasan, seperti rata-rata, total, median, dan lain sebagainya, untuk satu atau lebih variabel. Ini sering digunakan ketika kita ingin melihat data secara keseluruhan daripada secara individu.

6. **Transformasi Variabel (Variable Transformation):** Transformasi variabel melibatkan perubahan skala atau distribusi variabel. Contohnya, mengaplikasikan transformasi logaritma pada variabel untuk mengurangi skewness.

7. **Normalisasi dan Standarisasi (Normalization and Standardization):** Ini adalah proses mengubah nilai-nilai variabel sehingga mereka berada pada skala yang seragam. # Build Machine Learning Model

"Build Machine Learning Model" adalah suatu proses di mana kita mengembangkan model machine learning menggunakan algoritma tertentu untuk menyelesaikan suatu masalah atau tugas tertentu. Proses ini melibatkan beberapa tahapan umum, termasuk pengumpulan data, pemrosesan data, pemilihan model, pelatihan model, evaluasi model, dan penyesuaian model. Dalam konteks Support Vector Machine (SVM), mari jelaskan konsepnya secara singkat:

1. **Pengumpulan Data:**
   - Mengumpulkan dataset yang akan digunakan untuk melatih dan menguji model SVM. Dataset ini harus mencakup fitur-fitur yang relevan dengan masalah yang ingin dipecahkan dan label yang sesuai.

2. **Pemrosesan Data:**
   - Menangani missing values, melakukan normalisasi data, atau mengubah format data jika diperlukan. SVM sangat sensitif terhadap skala data, sehingga normalisasi umumnya diperlukan.

3. **Pemilihan Model:**
   - Memilih algoritma SVM dan konfigurasi yang sesuai untuk masalah yang dihadapi. SVM dapat digunakan untuk tugas klasifikasi (mengelompokkan data ke dalam kategori) dan regresi (mengestimasi nilai numerik).

4. **Pelatihan Model:**
   - Memisahkan dataset menjadi dua bagian: satu untuk pelatihan (training set) dan satu untuk pengujian (test set). Model SVM kemudian dilatih dengan menggunakan training set. Tujuan pelatihan adalah untuk menemukan hyperplane terbaik yang memisahkan antar kategori (dalam konteks klasifikasi).

5. **Evaluasi Model:**
   - Model dievaluasi menggunakan test set yang tidak pernah dilihat selama pelatihan. Metrik evaluasi yang umum digunakan untuk SVM meliputi akurasi, presisi, recall, F1-score, atau area di bawah kurva ROC, tergantung pada jenis masalah yang dihadapi.

6. **Penyesuaian Model:**
   - Berdasarkan hasil evaluasi, model dapat disesuaikan atau dioptimalkan untuk meningkatkan kinerja. Ini mungkin melibatkan penyesuaian parameter SVM, pemilihan fitur, atau strategi lainnya.

Penting untuk dicatat bahwa konsep-konsep di atas bersifat umum untuk proses pengembangan model machine learning dan dapat diterapkan pada berbagai algoritma, termasuk Support Vector Machine. SVM khususnya fokus pada mencari hyperplane terbaik yang memaksimalkan margin antara kategori dalam ruang fitur.

````python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split

# Assuming 'X' is your feature matrix, and 'y' is your target variable
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=99)

# Create an SVM classifier with specific parameters
svm_classifier = SVC(
    C=10,             # Regularization parameter
    kernel='rbf',      # Kernel type ('linear', 'poly', 'rbf', 'sigmoid', etc.)
    degree=3,          # Degree of the polynomial kernel function ('poly' only)
    gamma='scale',     # Kernel coefficient ('scale' uses 1 / (n_features * X.var()))
    coef0=0.0,         # Independent term in the kernel function ('poly' and 'sigmoid' only)
    shrinking=True,    # Whether to use the shrinking heuristic
    probability=False, # Whether to enable probability estimates
    tol=1e-3,          # Tolerance for stopping criterion
    random_state=42  # Seed for random number generation
)

# Train the SVM model
svm_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred_svm = svm_classifier.predict(X_test)
# Evaluate the SVM model
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print("Accuracy (SVM):", accuracy_svm)

precision_svm = precision_score(y_test, y_pred_svm, average='micro')
print("Precision (SVM):", precision_svm)

recall_svm = recall_score(y_test, y_pred_svm, average='micro')
print("Recall (SVM):", recall_svm)

f1_svm = f1_score(y_test, y_pred_svm, average='micro')
print("F1 Score (SVM):", f1_svm)

conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)
print("Confusion Matrix (SVM):")
print(conf_matrix_svm)

class_report_svm = classification_report(y_test, y_pred_svm)
print("Classification Report (SVM):")
print(class_report_svm)
````

# SMOTE, SMOTE - N, SMOTEEN

SMOTE, SMOTE - N, dan SMOTE-ENN merupakan teknik-teknik yang digunakan dalam pengolahan data untuk menangani ketidakseimbangan kelas (class imbalance) pada dataset. Masalah ketidakseimbangan kelas terjadi ketika jumlah sampel dalam satu kelas sangat tidak proporsional dengan jumlah sampel dalam kelas lainnya. Hal ini dapat menyebabkan model pembelajaran mesin cenderung memprediksi kelas mayoritas dengan akurasi yang tinggi, sementara kinerja pada kelas minoritas rendah.

Berikut adalah penjelasan singkat mengenai SMOTE, SMOTE - N, dan SMOTE-ENN:

1. **SMOTE (Synthetic Minority Over-sampling Technique):**
   - **Konsep:**
     SMOTE bertujuan untuk mengatasi ketidakseimbangan kelas dengan menciptakan sampel sintetis pada kelas minoritas. Ini dilakukan dengan membuat instance baru yang merupakan kombinasi dari instance minoritas yang sudah ada.
   - **Cara Kerja:**
     a. Pilih suatu instance dari kelas minoritas.
     b. Pilih k-nearest neighbors dari instance tersebut.
     c. Pilih satu dari tetangga-tetangga tersebut secara acak.
     d. Buat instance baru dengan mengambil perbedaan antara instance minoritas yang dipilih dan tetangga yang dipilih, kemudian tambahkan perbedaan tersebut pada instance awal.

2. **SMOTE - N (SMOTE for Nominal features):**
   - **Konsep:**
     SMOTE - N adalah modifikasi dari SMOTE yang dirancang untuk menangani atribut nominal dalam dataset.
   - **Cara Kerja:**
     a. Pilih suatu instance dari kelas minoritas.
     b. Pilih k-nearest neighbors dari instance tersebut.
     c. Pilih satu dari tetangga-tetangga tersebut secara acak.
     d. Buat instance baru dengan menambahkan perbedaan antara instance minoritas yang dipilih dan tetangga yang dipilih pada atribut nominal saja.

3. **SMOTE-ENN (SMOTE Edited Nearest Neighbors):**
   - **Konsep:**
     SMOTE-ENN adalah kombinasi dari SMOTE dan teknik penghapusan instance yang tidak relevan (Edited Nearest Neighbors, ENN). Tujuan utamanya adalah untuk meningkatkan kualitas sampel sintetis yang dihasilkan oleh SMOTE.
   - **Cara Kerja:**
     a. Gunakan SMOTE untuk membuat sampel sintetis pada kelas minoritas.
     b. Gunakan ENN untuk menghapus instance-instance dari kelas mayoritas yang salah klasifikasi.

Ketiga teknik ini membantu mengatasi masalah ketidakseimbangan kelas dengan menciptakan sampel sintetis atau dengan menghapus instance-instance yang tidak relevan. Ini membantu model pembelajaran mesin untuk mempelajari pola dari kelas minoritas dengan lebih baik, sehingga meningkatkan kinerja model pada kelas-kelas minoritas tersebut.

**Dataset Splitting Concepts**

Dataset splitting adalah proses membagi dataset menjadi subset yang berbeda untuk digunakan dalam pelatihan, validasi, dan pengujian model machine learning. Pembagian dataset ini penting untuk mengukur kinerja model secara obyektif dan mencegah overfitting, yaitu ketika model terlalu baik dalam mempelajari data pelatihan tetapi gagal saat diterapkan pada data baru.

Ada beberapa konsep utama dalam dataset splitting:

1. **Training Set (Dataset Pelatihan):** Ini adalah bagian dari dataset yang digunakan untuk melatih model. Model menggunakan data ini untuk belajar dan menyesuaikan parameter internalnya agar dapat membuat prediksi yang akurat.

2. **Validation Set (Dataset Validasi):** Bagian ini digunakan untuk mengevaluasi kinerja model selama pelatihan. Dengan menggunakan dataset validasi, Anda dapat mengukur sejauh mana model dapat menggeneralisasi pola yang ditemukan dalam data pelatihan ke data baru. Ini membantu dalam penyesuaian parameter model dan pencegahan overfitting.

3. **Test Set (Dataset Pengujian):** Setelah melatih model dan menyesuaikan parameter menggunakan dataset pelatihan dan validasi, test set digunakan untuk mengukur kinerja akhir model. Test set seharusnya tidak digunakan selama proses pelatihan atau validasi untuk memastikan evaluasi model bersifat independen.

4. **Cross-Validation:** Ini adalah teknik yang melibatkan pembagian dataset menjadi beberapa bagian (fold), di mana model dilatih pada beberapa kombinasi subset pelatihan dan diuji pada subset lainnya. Ini membantu menghasilkan estimasi yang lebih stabil tentang kinerja model daripada pembagian dataset yang tunggal.

5. **Holdout Method:** Ini adalah metode sederhana di mana dataset dibagi menjadi dua bagian besar: satu untuk pelatihan dan validasi, dan yang lainnya untuk pengujian. Metode ini dapat digunakan ketika dataset cukup besar.

6. **Stratified Sampling:** Pembagian dataset dilakukan sedemikian rupa sehingga distribusi kelas atau label yang ada dalam dataset dijaga pada setiap subset. Ini membantu memastikan bahwa setiap subset mencakup representasi yang seimbang dari kelas atau label yang ada.

Penting untuk mencatat bahwa ketika membagi dataset, pastikan bahwa data yang digunakan untuk pelatihan, validasi, dan pengujian mencerminkan karakteristik dan distribusi data secara keseluruhan. Ini membantu memastikan bahwa model yang dilatih dapat menghasilkan prediksi yang umumnya berlaku pada data baru.

**Training Process**

Training process dalam Support Vector Machine (SVM) melibatkan menemukan hyperplane optimal yang memisahkan dua kelas data. Berikut adalah langkah-langkah utama dalam pelatihan SVM:

1. **Pemilihan Kernel dan Parameter:**
   - Pilih jenis kernel yang sesuai. Kernel adalah fungsi matematis yang digunakan untuk mengubah ruang fitur asli menjadi ruang fitur yang lebih tinggi dimensi. Kernel umum melibatkan fungsi seperti linear, polynomial, radial basis function (RBF), atau sigmoid.
   - Tentukan parameter kernel, seperti C (parameter biaya) dan gamma (hanya berlaku untuk kernel RBF). Nilai-nilai ini memengaruhi kompleksitas model dan tingkat toleransi terhadap kesalahan pelatihan.

2. **Pemilihan Dataset:**
   - Pisahkan dataset menjadi dua bagian: satu untuk pelatihan dan satu untuk pengujian. Dataset pelatihan digunakan untuk menyesuaikan model, sedangkan dataset pengujian digunakan untuk mengevaluasi kinerja model yang dilatih.

3. **Normalisasi atau Standarisasi Data (Opsional):**
   - Beberapa kasus memerlukan normalisasi atau standarisasi data. Ini melibatkan penyesuaian nilai atribut sehingga mereka memiliki skala yang seragam, yang dapat meningkatkan kinerja SVM.

4. **Pelatihan Model:**
   - Model SVM dibangun dengan mencari hyperplane yang memaksimalkan margin antara dua kelas. Margin adalah jarak terpendek dari hyperplane ke titik terdekat dari kedua kelas.
   - SVM mencari hyperplane yang memenuhi dua persyaratan:
      - Memisahkan dua kelas dengan benar.
      - Maksimalkan margin antara dua kelas.

5. **Penanganan Ketidakseimbangan Kelas (Opsional):**
   - Jika dataset memiliki ketidakseimbangan kelas (jumlah sampel yang signifikan antara kelas satu dan kelas lainnya), teknik seperti oversampling atau undersampling dapat diterapkan untuk menangani masalah ini.

6. **Validasi Model (Opsional):**
   - Setelah melatih model, gunakan dataset validasi untuk menyesuaikan parameter dan mengoptimalkan kinerja model. Ini membantu mencegah overfitting pada dataset pelatihan.

7. **Evaluasi Model:**
   - Gunakan dataset pengujian yang terpisah untuk mengevaluasi kinerja model SVM. Metrik evaluasi umum termasuk akurasi, presisi, recall, dan F1-score.

8. **Tuning Parameter (Opsional):**
   - Sesuaikan parameter model, jika perlu, berdasarkan hasil evaluasi. Ini dapat melibatkan penggunaan teknik seperti cross-validation untuk menemukan parameter yang optimal.

Proses ini berfokus pada pencarian hyperplane yang efektif memisahkan dua kelas dalam ruang fitur. SVM juga dapat diadaptasi untuk penanganan masalah klasifikasi dengan lebih dari dua kelas menggunakan metode seperti one-vs-one atau one-vs-all.

**Evaluasi Model Support Vector Machine**

Evaluasi model Support Vector Machine (SVM) melibatkan sejumlah metrik dan teknik untuk mengukur kinerja dan keefektifan model tersebut dalam melakukan klasifikasi atau regresi. Berikut adalah beberapa konsep yang terkait dengan evaluasi model SVM:

1. **Confusion Matrix:**
   - True Positive (TP): Jumlah sampel positif yang diklasifikasikan dengan benar.
   - True Negative (TN): Jumlah sampel negatif yang diklasifikasikan dengan benar.
   - False Positive (FP): Jumlah sampel negatif yang salah diklasifikasikan sebagai positif.
   - False Negative (FN): Jumlah sampel positif yang salah diklasifikasikan sebagai negatif.

2. **Akurasi (Accuracy):**
   - Mengukur sejauh mana model SVM dapat mengklasifikasikan dengan benar seluruh sampel, dinyatakan dalam persentase.
   - \( \text{Akurasi} = \frac{TP + TN}{TP + TN + FP + FN} \)

3. **Presisi (Precision):**
   - Mengukur ketepatan model dalam mengklasifikasikan sampel positif.
   - \( \text{Presisi} = \frac{TP}{TP + FP} \)

4. **Recall (Sensitivitas atau Recall):**
   - Mengukur sejauh mana model dapat mendeteksi atau mengenali semua sampel positif yang sebenarnya.
   - \( \text{Recall} = \frac{TP}{TP + FN} \)

5. **F1-Score:**
   - Menggabungkan presisi dan recall menjadi satu metrik, berguna ketika keseimbangan antara presisi dan recall penting.
   - \( \text{F1-Score} = 2 \times \frac{\text{Presisi} \times \text{Recall}}{\text{Presisi} + \text{Recall}} \)

6. **Area di Bawah Kurva Receiver Operating Characteristic (AUC-ROC):**
   - Mengukur kemampuan model untuk membedakan antara kelas positif dan negatif dengan memplot tingkat true positive terhadap tingkat false positive pada berbagai threshold.
   - Nilai AUC-ROC mendekati 1 menunjukkan kinerja yang baik.

7. **Validation and Cross-Validation:**
   - Validasi model menggunakan data yang tidak digunakan selama pelatihan untuk mengukur kinerja pada data baru.
   - Cross-validation melibatkan pembagian data menjadi subset dan melakukan pelatihan dan pengujian beberapa kali untuk mendapatkan estimasi yang lebih stabil terhadap kinerja model.

8. **Hyperparameter Tuning:**
   - Evaluasi model SVM juga melibatkan penyetelan hyperparameter, seperti parameter kernel dan parameter cost (C), untuk meningkatkan kinerja model.

Penting untuk memahami bahwa evaluasi model SVM harus mencakup analisis yang komprehensif untuk memastikan bahwa model tersebut tidak hanya berkinerja baik pada data pelatihan tetapi juga dapat menggeneralisasi dengan baik pada data baru.

## Uji Data Baru

Uji Data Baru atau pengujian pada model Support Vector Machine (SVM) merupakan bagian penting dalam evaluasi kinerja model. Proses ini bertujuan untuk mengukur sejauh mana model yang telah dilatih mampu menggeneralisasi pola yang telah dipelajari dari data latih terhadap data yang belum pernah dilihat sebelumnya.

Berikut adalah penjelasan umum tentang Uji Data Baru pada pengujian model Support Vector Machine (SVM):

1. **Data Uji (Testing Data):** Sebelum melibatkan model SVM dalam uji data baru, perlu memisahkan data menjadi dua bagian utama: data latih (training data) dan data uji (testing data). Data latih digunakan untuk melatih model, sedangkan data uji digunakan untuk mengukur kinerja model di luar sampel data yang telah dilihat sebelumnya.

2. **Proses Pelatihan (Training Process):** SVM merupakan algoritma pembelajaran yang membangun sebuah hyperplane untuk memisahkan kelas-kelas dalam data latih. Hyperplane ini ditentukan berdasarkan vektor pendukung (support vectors) yang memiliki jarak terdekat dengan batas keputusan. Proses pelatihan berfokus pada menemukan hyperplane yang optimal.

3. **Pengujian Model dengan Data Uji Baru:**
   - Setelah model SVM dilatih dengan data latih, langkah berikutnya adalah menguji model menggunakan data uji yang belum pernah dilihat sebelumnya.
   - Data uji diberikan kepada model, dan prediksi dibuat untuk setiap sampel dalam data uji.
   - Hasil prediksi ini kemudian dibandingkan dengan label sebenarnya dari data uji untuk mengukur sejauh mana model dapat memprediksi dengan benar.

4. **Metrik Evaluasi Kinerja:**
   - Beberapa metrik evaluasi umum yang digunakan untuk mengukur kinerja model SVM pada data uji baru melibatkan perhitungan akurasi, presisi, recall, F1-score, dan matriks konfusi.
   - **Akurasi** mengukur sejauh mana model dapat membuat prediksi yang benar secara keseluruhan.
   - **Presisi dan recall** memberikan wawasan lebih rinci tentang kualitas prediksi model, terutama dalam konteks kelas-kelas yang tidak seimbang.
   - **F1-score** memberikan keseimbangan antara presisi dan recall.

5. **Penyetelan Hyperparameter (Optional):** Kadang-kadang, dilakukan penyetelan ulang terhadap hyperparameter model SVM untuk meningkatkan kinerja pada data uji. Ini dapat melibatkan menggunakan teknik seperti validasi silang untuk menemukan kombinasi hyperparameter yang optimal.

Dengan melakukan uji data baru, kita dapat menilai sejauh mana model SVM mampu menggeneralisasi pola yang telah dipelajari selama pelatihan dan seberapa baik model tersebut dapat berperforma pada data yang belum pernah dilihat sebelumnya.
